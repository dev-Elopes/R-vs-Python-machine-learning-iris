```python
import pandas as pd
from sklearn.model_selection import train_test_split


# Variavel para acessar a base 
caminho_arquivo = 'iris.data' 
# variavel com os nomes das colunas (não contém na base original)
nomes_colunas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
# setamos os nomes da colunas no arquivo da base
df_iris = pd.read_csv(caminho_arquivo, names=nomes_colunas)

# agora fazemos a divisão de teste e treino, com um embaralhamento das linhas e um seed que 
# define o embaralhamento padrão(evita diferentes versões da divisão),usaremos uma partição de 
# 30% para testes e o restante para treino, adicionamos uma estratificação para garantir que as 
# partições contenham o mesmo numero de linhas de cada classe
df_treino, df_teste = train_test_split(
    df_iris, 
    test_size=0.30, 
    random_state=42,     
    shuffle=True,         # Aplica o embaralhamento completo das linhas
    stratify=df_iris['class'] # estratifica baseado na coluna classe
)

# exporta os dois arquivos para uso dos modelos
df_treino.to_csv('iris_treino.csv', index=False)
df_teste.to_csv('iris_teste.csv', index=False)

print("Tratamento concluído!")
print(f"Instâncias de Treino: {len(df_treino)} | Instâncias de Teste: {len(df_teste)}")
```
